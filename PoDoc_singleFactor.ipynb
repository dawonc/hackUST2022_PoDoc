```python
import pandas as pd
import io
```

```python
from matplotlib import pyplot as plt
```

```python
dawon = pd.read_csv('dawondata.csv')
dowon = pd.read_csv('dowondata.csv')
minyoung = pd.read_csv('minyoungdata.csv')
hyojun = pd.read_csv('hyojundata.csv')
changmo = pd.read_csv('changmodata.csv')
changmo
```
