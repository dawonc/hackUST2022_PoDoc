```python
# df = pd.DataFrame(0, index=np.arange(len(dowon)), columns=17)
corr_dataset = pd.DataFrame(np.zeros((len(dowon), 15)))
corr_dataset
```

```python
for i in range(2,7):
    i_max = np.amax(dowon.iloc[:,i])
    i_min = np.amin(dowon.iloc[:,i])
    i_avg = (i_max + i_min)/2
    
    k = 0
    for value in dowon.iloc[:,i]:
        if (value > i_avg):
            corr_dataset[i-2][k] = 1
            k += 1
        else:
            corr_dataset[i-2][k] = 0
            k += 1            
            
corr_dataset
```

```python
j = 5
col = 5

for i in range(0, 4):
    for n in range(1,j):
        for k in range(len(corr_dataset[i])):            
            if (corr_dataset[i][k] == 1 and corr_dataset[i+n][k] == 1):
                corr_dataset[col][k] = 1
            elif (corr_dataset[i][k] == 1 and corr_dataset[i+n][k] == 0):
                corr_dataset[col][k] = 2
            elif (corr_dataset[i][k] == 0 and corr_dataset[i+n][k] == 1):
                corr_dataset[col][k] = 3
            elif (corr_dataset[i][k] == 0 and corr_dataset[i+n][k] == 0):
                corr_dataset[col][k] = 4
        col += 1
    j -= 1


corr_dataset 
```

```python
highSQ = pd.DataFrame(np.zeros((1, 10)))

for i in range(5, 15):
    sum = 0
    num = 0
    for k in range(len(corr_dataset[i])):
        if (dowon.iloc[k,1] == 4 or dowon.iloc[k,1] == 5):
            sum += corr_dataset[i][k]
            num += 1
    print(sum)
    print(num)
    highSQ[i-5] = sum / num

highSQ
```
