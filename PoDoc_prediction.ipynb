```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("dowondata.csv")
```

```python
print(data)
print(' ')
max_number=(24*3600)
list_avg_error=[]
list_avg_pred=[]
list_pred=[]
predu=[]
lcategory=0
category=['Daily Usage', 'Sleep Duration', 'Steps', 'Exercise Duration', 'Day of the Week']

for each_category in range (2,7):
    x = data.iloc[:,each_category:each_category+1].values
    y = data.iloc[:,1].values
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
   
    #feature scaling
    from sklearn.preprocessing import StandardScaler
    sc= StandardScaler()
    x_train= sc.fit_transform(x_train)
    x_test = sc.fit_transform(x_test)
    
    from sklearn.model_selection import KFold
    folds = KFold(n_splits=5,random_state=0,shuffle=True).split(x,y)
    
    q=0 #for calculating error
    r=0 #for calculating prediction
    
    each_category_number=each_category-2
    print(each_category_number+1,'.',category[each_category_number])
    if each_category_number ==0 or each_category_number ==1 or each_category_number ==3:   #units 'in hours'
        unit=' in hours'
    else:
        unit=''
    cc=float(input("Input of current "+str(category[each_category_number])+str(unit)+"?"))
    if each_category_number ==0 or each_category_number ==1 or each_category_number ==3:    #Convert seconds to hours
        cc=cc*3600
    print(' ')
    ccc=cc #for calculation prediction based on the current INPUT
    predu.append(cc)
    print('->Calculation with kfold')
    for trained, tested in folds:
        x_train = x[trained]
        y_train = y[trained]
        x_test = x[tested]
        y_test = y[tested]
        
        from sklearn.linear_model import LinearRegression
        regressor = LinearRegression()
        regressor.fit(x_train, y_train)
        y_pred = regressor.predict(x_test)
        np.set_printoptions(precision=2)
        #print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_pred),1)),1))
        p=0
        for n in range (len(y_pred)):
            p+=(y_pred[n]-y_test[n])**2
        print("Prediction Error is",np.sqrt(p)/(len(y_pred)))
        q+=np.sqrt(p)/(len(y_pred))  
    
        r+=regressor.predict([[predu[each_category_number]]])
        print("Predicted Satisfactory Level for",category[each_category_number],"with the value of", \
              predu[each_category_number],regressor.predict([[predu[each_category_number]]]))
        list_pred.append(regressor.predict([[predu[each_category_number]]]))

        if each_category ==3 or each_category ==4 or each_category ==5:   
        #example of recommendation for steps and exercise time
            if (r/5)<3:
                if regressor.predict([[ccc+60]]) > regressor.predict([[ccc]]):
                    while regressor.predict([[ccc]])<3:
                        ccc+=60
    #SUMMARY             
    print( " ")
    print(' ')
    print("SUMMARY FOR",category[each_category_number])
    print("Average of Predicted Satisfactory Level is ",r/5)
    list_avg_pred.append(r/5)
    list_avg_error.append(q/5)
    print("Average of Prediction Error of our calculation is",list_avg_error[each_category_number])
    print(" ")
    print(" ")
    print(" ")
    #RECOMMENDATION
    if each_category ==5:   #for exercise
        if (r/5)<3:
            print(" ")
            print('*******recommendation******:','With your current status, you satisfactory level will be less than 3')
            print(  "{:.2f}".format(ccc/3600), 'hours is recommended to reach at least 3, so you still need ',  \
                  "{:.2f}".format(ccc/3600-cc/3600),'hours')
            print(" ")
    if each_category ==4:   # for steps
        if (r/5)<3:
            print(" ")
            print('*******recommendation******:','With your current status, you satisfactory level will be less than 3')
            print(  "{:.2f}".format(ccc), 'steps is recommended to reach at least 3, so you still need ',   \
                  "{:.2f}".format(ccc-cc),'steps')
            print(" ")
    if each_category ==3:   # for sleep
        if (r/5)<3:
            print(" ***************")
            print('RECOMMENDATION:','With your current status, you satisfactory level will be less than 3')
            print(  "{:.2f}".format(ccc/3600), 'hours is recommended to reach at least 3, so you still need ',   \
                  "{:.2f}".format(ccc/3600-cc/3600),'hours')
            print("*************** ")
            print(' ')
    ccc=cc
    if list_avg_error[each_category_number] < max_number:
        max_number = list_avg_error[each_category_number]
        lcategory =category[each_category_number]
        lcategory_number =int(each_category_number)
    else:
        continue    
print(" ")
print(" ")
print(" ")

print("Out of the categories, ",lcategory,"has the biggest effect to this user's Sleep Satisfactory with the best correlation")

print(' ')
```
